```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
%matplotlib inline
```

```python
#functional method
plt.plot(x,y,'g')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1)
           #(nofrows,noofcolomn,plotnoreffer)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
#obejct oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_xlabel('X LABEL')
axes1.set_ylabel('Y LABEL')
axes1.set_title('LARGER PLOT')

axes2 = fig.add_axes([0.18,0.5,0.4,0.35])
                    #([left,bottom.width,height])
axes2.plot(y,x)
axes2.set_xlabel('X LABEL')
axes2.set_ylabel('Y LABEL')
axes2.set_title('SMALLER PLOT')
```
